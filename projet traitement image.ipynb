```python
data_df = pd.read_csv('data.csv')
data_df = pd.DataFrame(data_df['path'])
```

```python
def resize_image(image_path, width, height):
    # Read the image from file
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    # Resize the image
    image = cv2.resize(image, (width, height), interpolation = cv2.INTER_LINEAR)

    return image
```

```python
def color_comparison(filepath, algo):
    
    res = []
    
    bool_val = True
    for i in range(len(data_df)):
        tmp = []
        #print(df.iloc[i].path)
        #print(filepath)
        val, bool_val = algo(filepath, data_df.iloc[i].path)
        tmp.append(data_df.iloc[i].path)
        tmp.append(abs(val))
        res.append(tmp)
    res = sorted(res, key = lambda x: x[1], reverse = bool_val)
    
    img_paths = [item[0] for item in res[:18]]
    rates = [item[1] for item in res[:18]]
    
    
    return img_paths, rates 

def texture_comparison(file_path , algo):
    
    res = []
    
    bool_val = True
    
    res , bool_val = algo(filepath, data_df)
    
    res = sorted(res, key = lambda x: x[1], reverse = bool_val)
    
    img_paths = [item[0] for item in res[:18]]
    rates = [item[1] for item in res[:18]]
    
    
    return img_paths , rates 

def shape_comparison(file_path , algo):
    
    res = []
    
    bool_val = True
    
    res , bool_val = algo(filepath, data_df)
    
    res = sorted(res, key = lambda x: x[1], reverse = bool_val)
    
    img_paths = [item[0] for item in res[:18]]
    rates = [item[1] for item in res[:18]]
    
    
    return img_paths , rates 
```

```python
def show_color_results(file_path, algo_choice):
    
    second_window = tk.Toplevel(window)
    second_window.title("Results Searching by color")
    second_window.geometry("1200x800")

    
    image_paths, captions = color_comparison(file_path, algo_choice)

    
    # Create a grid layout
    for row in range(3):
        for col in range(6):
            i = row*6+col
            frame = Frame(second_window)
            frame.grid(row=row, column=col, padx = 14, pady = 8)
            
            
            # Read the image from file and convert it to a PhotoImage object
            img = cv2.imread(image_paths[i])
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img)
            img = ImageTk.PhotoImage(image=img)

            # Add the image to the Frame
            label = Label(frame, image=img)
            label.image = img  # keep a reference to the image
            label.pack()

            # Add the caption to the Frame
            label = Label(frame, text= captions[i], font=("Helvetica", 12))
            label.pack(side = "bottom")

            
    
    # Run the Tkinter event loop
    second_window.mainloop()

def show_texture_results(file_path, algo_choice):
    
    second_window = tk.Toplevel(window)
    second_window.title("Results Searching by texture")
    second_window.geometry("1200x800")

    
    image_paths, captions = texture_comparison(file_path, algo_choice)
    
    # Create a grid layout
    for row in range(3):
        for col in range(6):
            i = row*6+col
            frame = Frame(second_window)
            frame.grid(row=row, column=col, padx = 14, pady = 8)
            
            
            # Read the image from file and convert it to a PhotoImage object
            img = cv2.imread(image_paths[i])
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img)
            img = ImageTk.PhotoImage(image=img)

            # Add the image to the Frame
            label = Label(frame, image=img)
            label.image = img  # keep a reference to the image
            label.pack()

            # Add the caption to the Frame
            label = Label(frame, text= captions[i], font=("Helvetica", 12))
            label.pack(side = "bottom")

            
    
    # Run the Tkinter event loop
    second_window.mainloop() 

def show_shape_results(file_path, algo_choice):
    
    second_window = tk.Toplevel(window)
    second_window.title("Results Searching by shape")
    second_window.geometry("1200x800")

    
    image_paths, captions = shape_comparison(file_path, algo_choice)
    
    # Create a grid layout
    for row in range(3):
        for col in range(6):
            i = row*6+col
            frame = Frame(second_window)
            frame.grid(row=row, column=col, padx = 14, pady = 8)
            
            
            # Read the image from file and convert it to a PhotoImage object
            img = cv2.imread(image_paths[i])
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img)
            img = ImageTk.PhotoImage(image=img)

            # Add the image to the Frame
            label = Label(frame, image=img)
            label.image = img  # keep a reference to the image
            label.pack()

            # Add the caption to the Frame
            label = Label(frame, text= captions[i], font=("Helvetica", 12))
            label.pack(side = "bottom")

            
    
    # Run the Tkinter event loop
    second_window.mainloop() 
    
    
    
    
    
  
```

```python
# color testing methods

def HSV(image_path, test_path):
    
    # read the image 
    
    
    base = resize_image(image_path, 220, 220)
    test = cv2.imread(test_path, cv2.IMREAD_COLOR)
    
    # convert to hsv
    hsv_base = cv2.cvtColor(base, cv2.COLOR_BGR2HSV)
    hsv_test = cv2.cvtColor(test, cv2.COLOR_BGR2HSV)

    h_bins = 180
    s_bins = 256
    v_bins = 256
    histSize = [h_bins, s_bins, v_bins ]
    ranges = [0, 180, 0, 256, 0, 256]
    channels = [0, 1, 2]

    hist_base = cv2.calcHist([hsv_base], channels, None, histSize, ranges, accumulate=False)
    cv2.normalize(hist_base, hist_base, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)
    hist_test = cv2.calcHist([hsv_test], channels, None, histSize, ranges, accumulate=False)
    cv2.normalize(hist_test, hist_test, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

    compare_method = cv2.HISTCMP_CORREL

    base_base = cv2.compareHist(hist_base, hist_test, compare_method)

    return base_base , True



def RGB(image_path, test_path):
    
    
    resized_image = transform.resize(io.imread(image_path), (220,220))

    # Read the image
    image1 = img_as_float(resized_image)

    image2 = img_as_float(io.imread(test_path))
    
    score = ssim(image1, image2,multichannel=True)
    
    return score , True


def MSE(image_path , test_path):
    
    # read the images with the same size 
    image1 = resize_image(image_path, 220, 220)
    image2 = cv2.imread(test_path)

    # convert the images to gray scale images 
    gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # Normalize the images to a common range of pixel values
    gray1 = gray1 / 255.0
    gray2 = gray2 / 255.0

    # Calculate the mean squared
    
    mse = np.mean((gray1 - gray2) ** 2)
        
    return mse , False
```

```python
# texture testing methods

def list_stat_cooccurance(df):
    n = df.shape[0]
    stat = []
    for i in range(n) :
        tmp_stat = []
        image = cv2.imread(df['path'].iloc[i])
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        result = feature.greycomatrix(image, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256)
        tmp_stat.append(feature.greycoprops(result,'contrast'))
        tmp_stat.append(feature.greycoprops(result,'correlation'))
        tmp_stat.append(feature.greycoprops(result,'dissimilarity'))
        tmp_stat.append(feature.greycoprops(result,'ASM'))
        stat.append(tmp_stat)
    return stat

def stat_cooccurance(image_path):
    tmp_stat = []
    image = resize_image(image_path, 220, 220)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    result = feature.greycomatrix(image, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256)
    tmp_stat.append(feature.greycoprops(result,'contrast'))
    tmp_stat.append(feature.greycoprops(result,'correlation'))
    tmp_stat.append(feature.greycoprops(result,'dissimilarity'))
    tmp_stat.append(feature.greycoprops(result,'ASM'))
    return tmp_stat

def cooccurance(image_path, df):
    PD = []
    statImg = stat_cooccurance(image_path)
    listStat = list_stat_cooccurance(df)
    n = len(listStat)
    m = len(listStat[0])
    for i in range(n):
        tmp_pd = [df['path'].iloc[i]]
        d = 0
        for j in range(m):
            m1 = sum(listStat[i][j][0])/len(listStat[i][j][0])
            m2 = sum(statImg[j][0])/len(statImg[j][0])
            d +=  (m2-m1)**2
        tmp_pd.append(d)
        PD.append(tmp_pd)
    return PD , False


def list_stat_LBP(df):
    n = df.shape[0]
    stat = []
    for i in range(n) :
        image = cv2.imread(df['path'].iloc[i])
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        result = feature.local_binary_pattern(image, 8,2, method='uniform')
        mode= st.mode(result[0])
        moy= result.mean()
        var= np.var((result[0]))
        stati = []
        stati.append(mode)
        stati.append(moy)
        stati.append(var)
        stat.append(stati)
    return stat
def stat_LBP(image_path):
    image = resize_image(image_path, 220, 220)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    result = feature.local_binary_pattern(image, 8,2, method='uniform')
    mode= st.mode(result[0])
    moy= result.mean()
    var= np.var((result[0]))
    stati = []
    stati.append(mode)
    stati.append(moy)
    stati.append(var)
    return stati

def LBP(image_path,df):
    a = list_stat_LBP(df)
    b = stat_LBP(image_path)
    
    n = len(a)
    m = len(b)
    D = []
    for i in range(n):
        tmp_pd = [df['path'].iloc[i]]
        d = 0
        for j in range(m):
            d+= (a[i][j]-b[j])**2
        tmp_pd.append(sqrt(d))
        D.append(tmp_pd)
    return D , False


```

```python
# shape testing methods 


def HU_distance(image_path, df) : 
    img1 = resize_image(image_path, 220, 220)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    ret, img1 = cv2.threshold(img1, 128, 255, cv2.THRESH_BINARY)
    img1 = cv2.normalize(img1, None, 0, 1, cv2.NORM_MINMAX)
    contours1, _ = cv2.findContours(img1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    hu1 = cv2.HuMoments(cv2.moments(contours1[0])).flatten()
    n = df.shape[0]
    distances = []
    for i in range(n) :
        tmp_df = [df['path'].iloc[i]]
        img2 = cv2.imread(df['path'].iloc[i], cv2.IMREAD_GRAYSCALE)
        ret, img2 = cv2.threshold(img2, 128, 255, cv2.THRESH_BINARY)
        img2 = cv2.normalize(img2, None, 0, 1, cv2.NORM_MINMAX)
        contours2, _ = cv2.findContours(img2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        hu2 = cv2.HuMoments(cv2.moments(contours2[0])).flatten()
        d2 = cv2.norm(hu1, hu2, cv2.NORM_L1)
        tmp_df.append(d2)
        distances.append(tmp_df)
    return distances , False


def Zernike(image_path, df):
    
    img = resize_image(image_path , 220, 220)
    img = img[:, :, 0]
    radius = 10
    zernike1 = mahotas.features.zernike_moments(img, radius)
    distances = []
    n = df.shape[0]
    for i in range(n) :
        imgs = cv2.imread(df['path'].iloc[i])
        # filtering image
        imgs = imgs[:, :, 0]
        # radius
        
        # computing zernike moments
        zernike2 = mahotas.features.zernike_moments(imgs, radius)
        distance = euclidean(zernike1, zernike2)
        distances.append([df['path'].iloc[i],  distance])
    
    
    
    return distances , False
```
